```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
mainData=pd.read_csv("C:\\Users\\rober\\Downloads\\IndeedBIHomework (1) (1) (1) (1) (1) (1) (2)\\BIHomework\\NEISS2014.csv")
diagCodes=pd.read_csv("C:\\Users\\rober\\\Downloads\\IndeedBIHomework (1) (1) (1) (1) (1) (1) (2)\\BIHomework\\DiagnosisCodes.csv")
```

What diagnosis had the highest hospitalization rate? 

```python
diagnosisMerge=pd.merge(mainData,diagCodes,how="left", left_on="diag",right_on="Code")
```

```python
hospitalised=diagnosisMerge[(diagnosisMerge['disposition']==4)]
```

```python
hospitalised
```

```python
countOfHospitalised=hospitalised.groupby('Diagnosis')['Diagnosis'].agg(['count']).reset_index()
```

```python
countOfHospitalised=countOfHospitalised.rename(columns={'count':'Hospitalised'})
```

```python
countOfTotal = diagnosisMerge.groupby('Diagnosis')['Diagnosis'].agg(['count']).reset_index()
```

```python
hospitalisedVsTotal = pd.merge(countOfHospitalised, countOfTotal, how='left', on='Diagnosis')
```

```python
hospitalisedVsTotal = hospitalisedVsTotal.rename(columns={'count_x':'Hospitalised', 'count_y':'Total'})
```

```python
countSorted = hospitalisedVsTotal.sort_values(by='Hospitalised', ascending=False)
```

```python
countSorted
```

```python
countSorted['Rate']=(countSorted['Hospitalised']/hospitalisedVsTotal['Total'])*100
```

```python
countSorted = countSorted.sort_values(by='Rate', ascending=False)
```

```python
top5 = countSorted.head(5)
```

```python
top5=top5.reset_index()
```

```python
top5
```

```python
top5['Diagnosis'][0]
```

```python
print('The Top hospitalisation rate is: '+ top5['Diagnosis'][0])
```

```python
#What diagnosis most often concluded with the individual leaving without being seen?
```

```python
leftWOSeen=diagnosisMerge[(diagnosisMerge['disposition']==6)]
```

```python
countWOSeen=leftWOSeen.groupby('Diagnosis')['Diagnosis'].agg(['count']).reset_index()
```

```python
countWOSeen=countWOSeen.rename(columns={'count':'not seen'})
```

```python
leftVsTotal=pd.merge(countWOSeen,countOfTotal,how="left", on="Diagnosis")
```

```python
leftVsTotal['rate']=(leftVsTotal['not seen']/leftVsTotal['count'])*100
```

```python
leftVsTotal = leftVsTotal.rename(columns={'rate':'rate %'})
```

```python
sortedLeftVTotal = leftVsTotal.sort_values(by='rate %', ascending=False).reset_index()
```

```python
sortedLeftVTotal.head()
```

```python
print('The diagnosis most associated with people leaving untreated is: '+sortedLeftVTotal['Diagnosis'][0])
```
